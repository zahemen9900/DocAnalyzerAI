```python
income_df = pd.concat({'Income': apple.financials.income.to_dataframe()})
income_df.head()
```

```python
cashflow_df = pd.concat({'Cashflows': apple.financials.cashflow.to_dataframe()})
cashflow_df.head()
```

```python
apple_financial_df = pd.concat([income_df, cashflow_df], axis=0).T
apple_financial_df
```
