 Other Methods 

```python
income_df = apple.financials.income.to_dataframe().T

cols = income_df.columns.tolist()
for col in cols:
    if col not in ['Basic (in dollars per share)', 'Diluted (in dollars per share)']:
        income_df[col] = income_df[col].astype(int)

income_df[['Basic (in dollars per share)', 'Diluted (in dollars per share)']] = income_df[['Basic (in dollars per share)', 'Diluted (in dollars per share)']].astype('Float32')

income_df.head()
```

```python
cashflow_df = apple.financials.cashflow.to_dataframe().T

for col in cashflow_df.columns.tolist():
    cashflow_df[col] = cashflow_df[col].astype(int)

cashflow_df.head()
```

```python
income_df.columns
```

```python
# Some feature engineering
income_df['Net_Sale_Growth_rate'] = (income_df['Net sales'] / income_df['Net sales'].shift(1) - 1) * 100
income_df['Net_Income_Growth_rate'] = (income_df['Net income'] / income_df['Net income'].shift(1) - 1) * 100
income_df.fillna(0, inplace=True)

# Drop less relevant columns
income_df =income_df.drop(['Products',
  'Services',
  'Products',
  'Services',
  'Basic (in shares)',
  'Diluted (in shares)']
  , axis=1)

income_df.head()
```

```python
cashflow_df = cashflow_df.drop(['Cash, cash equivalents, and restricted cash and cash equivalents, ending balances',
  'Share-based compensation expense',
  'Other',
  'Vendor non-trade receivables',
  'Other',
  'Payments for taxes related to net share settlement of equity awards',
  'Proceeds from/(Repayments of) commercial paper, net',
  'Other'], axis=1)

cashflow_df.head()
```

```python
apple_financial_df = pd.concat([income_df, cashflow_df], axis=1)
apple_financial_df
```
